```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

all_features = df.columns.drop(['Female', 'Year'])
train_data = df[df['Year'] <= 2018]
test_data = df[(df['Year'] >= 2019) & (df['Year'] <= 2022)]

X_train = train_data[all_features]
y_train_female = train_data['Female']

X_test = test_data[all_features]
y_test_female = test_data['Female']

# Training the models for female population
r_model_train_female = LinearRegression()
K_model_train_female = LinearRegression()

r_model_train_female.fit(X_train, y_train_female)
K_model_train_female.fit(X_train, y_train_female)

r_params_train_female = [r_model_train_female.intercept_] + list(r_model_train_female.coef_)
K_params_train_female = [K_model_train_female.intercept_] + list(K_model_train_female.coef_)

# Initial female population
P0_test_female = y_train_female.iloc[-1]

# Defining necessary functions
def compute_r_K_limited(X, r_params, K_params, r_min, r_max, K_min, K_max):
    r = np.clip(r_params[0] + sum(r_param * x for r_param, x in zip(r_params[1:], X)), r_min, r_max)
    K = np.clip(K_params[0] + sum(K_param * x for K_param, x in zip(K_params[1:], X)), K_min, K_max)
    return r, K

def logistic_growth(P, t, r, K):
    return r * P * (1 - P / K)

def euler_method_limited(P0, t, r_params, K_params, X, r_min, r_max, K_min, K_max):
    dt = t[1] - t[0]
    P = [P0]
    for i in range(1, len(t)):
        r, K = compute_r_K_limited(X, r_params, K_params, r_min, r_max, K_min, K_max)
        P_next = P[-1] + logistic_growth(P[-1], t[i-1], r, K) * dt
        P.append(P_next)
    return P

r_min, r_max = 0, 1
K_min, K_max = 1e5, 1e8  

t_corrected = np.linspace(0, len(test_data) - 1, len(test_data) * 250)

# Forecasting the female population
population_forecast_corrected_female = euler_method_limited(
    P0_test_female, 
    t_corrected, 
    r_params_train_female, 
    K_params_train_female, 
    X_test.iloc[-1].values, 
    r_min, 
    r_max, 
    K_min, 
    K_max
)

# Extracting predicted female population at corresponding time points
predicted_population_corrected_female = population_forecast_corrected_female[::len(t_corrected) // len(test_data)]

# Calculating MSE and R2 score for the prediction
mse_corrected_female = mean_squared_error(y_test_female, predicted_population_corrected_female)
r2_corrected_female = r2_score(y_test_female, predicted_population_corrected_female)

# Plotting the results
years_train = np.arange(2009, 2023)
population_full_train_female = np.concatenate((y_train_female.values, predicted_population_corrected_female))

plt.figure(figsize=(10, 6))
plt.plot(years_train, population_full_train_female, marker='o', linestyle='-', color='b', label='Прогнозируемое население женщин (обновленное)')
plt.plot(df['Year'], df['Female'], marker='x', linestyle='-', color='r', label='Реальное население женщин')
plt.xlabel('Год')
plt.ylabel('Население женщин')
plt.title('Прогноз и реальное население женщин (с использованием данных до 2018 года, обновленный)')
plt.legend()
plt.grid(True)
plt.show()

mse_corrected_female, r2_corrected_female
```

```python
def malthusian_model(P0, r, t):
    return P0 * np.exp(r * t)

# Calculate the initial population and r values for the female population
P0_female = y_train_female.iloc[0]
r_values_female = []

for i in range(1, len(y_train_female)):
    r = np.log(y_train_female.iloc[i] / y_train_female.iloc[i-1])
    r_values_female.append(r)

r_mean_female = np.mean(r_values_female)

# Predicting the female population using the Malthusian model
years_test = np.arange(2019, 2023) - 2018
predicted_population_malthus_female = malthusian_model(y_train_female.iloc[-1], r_mean_female, years_test)

# Calculate MSE and R2 score for the Malthusian model prediction for female population
mse_malthus_female = mean_squared_error(y_test_female, predicted_population_malthus_female)
r2_malthus_female = r2_score(y_test_female, predicted_population_malthus_female)

# Plotting the results for the female population
years_full = np.arange(2009, 2023)
population_full_malthus_female = np.concatenate((y_train_female.values, predicted_population_malthus_female))

plt.figure(figsize=(10, 6))
plt.plot(years_full, population_full_malthus_female, marker='o', linestyle='-', color='b', label='Прогнозируемое население женщин (модель Мальтуса)')
plt.plot(df['Year'], df['Female'], marker='x', linestyle='-', color='r', label='Реальное население женщин')
plt.xlabel('Год')
plt.ylabel('Население женщин')
plt.title('Прогноз и реальное население женщин (с использованием данных до 2018 года, модель Мальтуса)')
plt.legend()
plt.grid(True)
plt.show()

mse_malthus_female, r2_malthus_female
```

```python
def verhulst_model(P0, r, K, t):
    return K * P0 * np.exp(r * t) / (K + P0 * (np.exp(r * t) - 1))

# Calculate the initial population and r values for the female population
P0_female = y_train_female.iloc[0]
r_values_female = []

for i in range(1, len(y_train_female)):
    r = np.log(y_train_female.iloc[i] / y_train_female.iloc[i-1])
    r_values_female.append(r)

r_mean_female = np.mean(r_values_female)

# Carrying capacity for the female population
K_female = y_train_female.max() * 1.5  

# Predicting the female population using the Verhulst model
predicted_population_verhulst_female = verhulst_model(y_train_female.iloc[-1], r_mean_female, K_female, years_test)

# Calculate MSE and R2 score for the Verhulst model prediction for female population
mse_verhulst_female = mean_squared_error(y_test_female, predicted_population_verhulst_female)
r2_verhulst_female = r2_score(y_test_female, predicted_population_verhulst_female)

# Plotting the results for the female population
population_full_verhulst_female = np.concatenate((y_train_female.values, predicted_population_verhulst_female))

plt.figure(figsize=(10, 6))
plt.plot(years_full, population_full_verhulst_female, marker='o', linestyle='-', color='b', label='Прогнозируемое население женщин (модель Вера Гольдштейна)')
plt.plot(df['Year'], df['Female'], marker='x', linestyle='-', color='r', label='Реальное население женщин')
plt.xlabel('Год')
plt.ylabel('Население женщин')
plt.title('Прогноз и реальное население женщин (с использованием данных до 2018 года, модель Вера Гольдштейна)')
plt.legend()
plt.grid(True)
plt.show()

mse_verhulst_female, r2_verhulst_female
```
