```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('demography_with_age.csv')

df
```

```python
df = df[9:]
df
```

```python
df = df[df['Year'] != 2024]
df['Population'] = df['Population'].astype(str).str.replace(' ', '').astype(int)
df['Male'] = df['Male'].astype(str).str.replace(' ', '')
df['Female'] = df['Female'].astype(str).str.replace(' ', '')


columns = ['Number of Death(Male)', 'Number of Death(Female)', 'Total Number of Death',
           'Number of Arrivals', 'Number of Deaprtures', 'Migration Difference',
           'Total Migration Difference']

# Iterate over the list of column names
for column in columns:
    # Apply string manipulation and conversion for each column
    df[column] = df[column].astype(str).str.replace(' ', '').str.replace('\xa0', '').astype(float)

# Now, the specified columns have been modified to integers without spaces or non-breaking spaces
df
```

```python
df.info()
```

```python
df['Male'] = pd.to_numeric(df['Male'], errors='coerce')
df['Female'] = pd.to_numeric(df['Female'], errors='coerce')
```

```python
coef_columns = ['Birth Coef', 'Death Coef', 'Death Coef(Male)', 'Death Coef(Female)', 'Death Coef(Child)', 'Death Coef(Boys)', 'Death Coef(Girls)']
for coef in coef_columns:
    df[coef] = df[coef].str.replace(',', '.').astype(float)
df
```

```python
from sklearn.linear_model import LinearRegression
df['Year'] = df['Year'].astype(int)
df_clean = df[['Year', 'Birth Coef']].dropna()
X = df_clean['Year'].values.reshape(-1, 1)
y = df_clean['Birth Coef'].values

model = LinearRegression()
model.fit(X, y)


year_2024 = np.array([[2024]])
birth_coef_2024 = model.predict(year_2024)

print(f"Predicted Birth Coefficient for 2024: {birth_coef_2024[0]}")
```

```python
def cohort_component_model(df, birth_coef_2024):
    df_next_year = df.iloc[-1:].copy()
    df_next_year['Year'] += 1

   
    df_next_year['Male'] = df_next_year['Male'] * (df_next_year['Birth Coef'] - df_next_year['Death Coef(Male)']) + df_next_year['Total Migration Difference'] * 0.5
    df_next_year['Female'] = df_next_year['Female'] * (df_next_year['Birth Coef'] - df_next_year['Death Coef(Female)']) + df_next_year['Total Migration Difference'] * 0.5

    total_births = birth_coef_2024 * df_next_year['Female_Age(20-30)'].sum()
    births_male = total_births * 0.5
    births_female = total_births * 0.5

    
    df_next_year['Male_Age(80+)'] = df.iloc[-1]['Male_Age(70-80)']
    df_next_year['Male_Age(70-80)'] = df.iloc[-1]['Male_Age(60-70)']
    df_next_year['Male_Age(60-70)'] = df.iloc[-1]['Male_Age(50-60)']
    df_next_year['Male_Age(50-60)'] = df.iloc[-1]['Male_Age(40-50)']
    df_next_year['Male_Age(40-50)'] = df.iloc[-1]['Male_Age(30-40)']
    df_next_year['Male_Age(30-40)'] = df.iloc[-1]['Male_Age(20-30)']
    df_next_year['Male_Age(20-30)'] = df.iloc[-1]['Male_Age(10-20)']
    df_next_year['Male_Age(10-20)'] = df.iloc[-1]['Male_Age(0-10)']
    df_next_year['Male_Age(0-10)'] = births_male

    df_next_year['Female_Age(80+)'] = df.iloc[-1]['Female_Age(70-80)']
    df_next_year['Female_Age(70-80)'] = df.iloc[-1]['Female_Age(60-70)']
    df_next_year['Female_Age(60-70)'] = df.iloc[-1]['Female_Age(50-60)']
    df_next_year['Female_Age(50-60)'] = df.iloc[-1]['Female_Age(40-50)']
    df_next_year['Female_Age(40-50)'] = df.iloc[-1]['Female_Age(30-40)']
    df_next_year['Female_Age(30-40)'] = df.iloc[-1]['Female_Age(20-30)']
    df_next_year['Female_Age(20-30)'] = df.iloc[-1]['Female_Age(10-20)']
    df_next_year['Female_Age(10-20)'] = df.iloc[-1]['Female_Age(0-10)']
    df_next_year['Female_Age(0-10)'] = births_female

    return df_next_year


df_2024 = cohort_component_model(df, birth_coef_2024[0])

print("Projected DataFrame for the year 2024:")
df_2024
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
features = df[['Year', 'Birth Coef', 'Death Coef(Male)', 'Death Coef(Female)', 'Total Migration Difference']]
target_male = df['Male']
target_female = df['Female']


X_train, X_test, y_train_male, y_test_male = train_test_split(features, target_male, test_size=0.2, random_state=42)
_, _, y_train_female, y_test_female = train_test_split(features, target_female, test_size=0.2, random_state=42)


xgb_model_male = XGBRegressor(objective='reg:squarederror', n_estimators=100)
xgb_model_male.fit(X_train, y_train_male)

xgb_model_female = XGBRegressor(objective='reg:squarederror', n_estimators=100)
xgb_model_female.fit(X_train, y_train_female)


features_2024 = pd.DataFrame({
    'Year': [2024],
    'Birth Coef': [df['Birth Coef'].mean()],  
    'Death Coef(Male)': [df['Death Coef(Male)'].mean()],  
    'Death Coef(Female)': [df['Death Coef(Female)'].mean()],  
    'Total Migration Difference': [df['Total Migration Difference'].mean()]  
})

predicted_male_2024 = xgb_model_male.predict(features_2024)
predicted_female_2024 = xgb_model_female.predict(features_2024)
overall_population = predicted_female_2024 + predicted_male_2024
print(f"Predicted Population for 2024: {overall_population}")
print(f"Predicted Male Population for 2024: {predicted_male_2024[0]}")
print(f"Predicted Female Population for 2024: {predicted_female_2024[0]}")


y_pred_male = xgb_model_male.predict(X_test)
y_pred_female = xgb_model_female.predict(X_test)

rmse_male = np.sqrt(mean_squared_error(y_test_male, y_pred_male))
rmse_female = np.sqrt(mean_squared_error(y_test_female, y_pred_female))

print(f"RMSE for Male Population Prediction: {rmse_male}")
print(f"RMSE for Female Population Prediction: {rmse_female}")


output_df = pd.DataFrame({
    'Year': [2024],
    'Male': predicted_male_2024,
    'Female': predicted_female_2024
})
```
