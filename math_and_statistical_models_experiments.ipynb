```python
import matplotlib.pyplot as plt
df_until_2015 = df[df['Year'] <= 2015]


initial_population_2015 = df_until_2015['Population'].iloc[-1]


birth_coef_2015 = df_until_2015['Total number of Birth()']
death_coef_2015 = df_until_2015['Total Number of Death']
migration_diff_2015 = df_until_2015['Total Migration Difference']


years_pred = np.arange(2016, 2025)
predicted_population_2015_2023 = [initial_population_2015]

for year in years_pred:
    birth = np.interp(year, df_until_2015['Year'], birth_coef_2015)
    death = np.interp(year, df_until_2015['Year'], death_coef_2015)
    migration = np.interp(year, df_until_2015['Year'], migration_diff_2015)
    next_population = predicted_population_2015_2023[-1] + birth - death + migration
    predicted_population_2015_2023.append(next_population)

df_pred_2015_2023 = pd.DataFrame({
    'Year': np.arange(2015, 2025),
    'Predicted Population': [initial_population_2015] + predicted_population_2015_2023[:-1]
})


print(df_pred_2015_2023)


plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Population'], label='Actual Population', marker='o')
plt.plot(df_pred_2015_2023['Year'], df_pred_2015_2023['Predicted Population'], label='Predicted Population (2015-2023)', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.title('Population Prediction (2015-2023)')
plt.show()
```

```python
from scipy.integrate import odeint
from scipy.interpolate import interp1d
r = 0.15
K = 2500000


initial_population = df['Population'].iloc[0]


birth = df['Total number of Birth()']
death = df['Total Number of Death']
migration_diff = df['Total Migration Difference']


interp_birth = interp1d(df['Year'], birth, fill_value="extrapolate")
interp_death = interp1d(df['Year'], death, fill_value="extrapolate")
interp_migration = interp1d(df['Year'], migration_diff, fill_value="extrapolate")


def population_model(P, t, r, K):
    B = interp_birth(t)
    D = interp_death(t)
    M = interp_migration(t)
    dPdt = r * P * (1 - P / K) + B - D + M
    return dPdt


t = np.linspace(2015, 2025, 100)
P = odeint(population_model, initial_population, t, args=(r, K))

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Population'], label='Actual Population', marker='o')
plt.plot(t, P, label='Predicted Population', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Prediction Using Differential Equations')
plt.show()
```

<h1>Logistic Growth Function</h1>


<h3>Training set in till 2020</h3>
<h4>Predict for 2030</h4>

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assuming df_with_age is already defined
population_data = df_with_age[['Year', 'Population']]

# Set the Year column as the index
population_data.set_index('Year', inplace=True)

# Define the logistic growth function
def logistic_growth(t, P0, r, K):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Extract the years and population data for curve fitting
years = np.array(population_data.index)
population = np.array(population_data['Population'])

# Split data into train and test sets
train_years = years[years <= 2020]
test_years = years[years > 2020]
train_population = population[years <= 2020]
test_population = population[years > 2020]

# Fit the logistic growth model to the training data
params, _ = curve_fit(logistic_growth, train_years - train_years[0], train_population, p0=[train_population[0], 0.1, max(train_population) * 2])

# Extract the estimated parameters
P0, r, K = params

# Forecast the population using the logistic growth model up to 2030
forecast_years = np.arange(train_years[0], 2031)
forecast_population = logistic_growth(forecast_years - train_years[0], P0, r, K)

# Create a DataFrame for the forecast results
forecast_df_logistic = pd.DataFrame({'Year': forecast_years, 'Forecasted Population': forecast_population})

pred_train_population = logistic_growth(train_years - train_years[0], P0, r, K)
pred_test_population = logistic_growth(test_years - train_years[0], P0, r, K)

# Calculate metrics
train_mae = mean_absolute_error(train_population, pred_train_population)
train_mse = mean_squared_error(train_population, pred_train_population)
train_r2 = r2_score(train_population, pred_train_population)

test_mae = mean_absolute_error(test_population, pred_test_population)
test_mse = mean_squared_error(test_population, pred_test_population)
test_r2 = r2_score(test_population, pred_test_population)

# Print metrics
print(f'Train MAE: {train_mae:.2f}')
print(f'Train MSE: {train_mse:.2f}')
print(f'Train R²: {train_r2:.2f}')
print(f'Test MAE: {test_mae:.2f}')
print(f'Test MSE: {test_mse:.2f}')
print(f'Test R²: {test_r2:.2f}')


# Plot the actual data and the logistic growth forecast
plt.figure(figsize=(10, 6))
plt.plot(train_years, train_population, marker='o', linestyle='-', label='Train Population')
plt.plot(test_years, test_population, marker='o', linestyle='-', label='Test Population')
plt.plot(forecast_df_logistic['Year'], forecast_df_logistic['Forecasted Population'], marker='o', linestyle='--', color='red', label='Forecasted Population (Logistic Model)')
plt.title('Population Forecast Using Logistic Growth Model')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()

# Save the forecast results to a CSV file
forecast_df_logistic.to_csv('forecasted_population.csv', index=False)

# Print the forecast DataFrame
print(forecast_df_logistic)
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assuming df_with_age is already defined
population_data = df[['Year', 'Population']]

# Set the Year column as the index
population_data.set_index('Year', inplace=True)

# Define the logistic growth function
def logistic_growth(t, P0, r, K):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Extract the years and population data for curve fitting
years = np.array(population_data.index)
population = np.array(population_data['Population'])

# Split data into train and test sets
train_years = years[years <= 2015]
test_years = years[years > 2015]
train_population = population[years <= 2015]
test_population = population[years > 2015]

# Fit the logistic growth model to the training data
params, _ = curve_fit(logistic_growth, train_years - train_years[0], train_population, p0=[train_population[0], 0.1, max(train_population) * 2])

# Extract the estimated parameters
P0, r, K = params

# Forecast the population using the logistic growth model up to 2030
forecast_years = np.arange(train_years[0], 2031)
forecast_population = logistic_growth(forecast_years - train_years[0], P0, r, K)

# Create a DataFrame for the forecast results
forecast_df_logistic = pd.DataFrame({'Year': forecast_years, 'Forecasted Population': forecast_population})

pred_train_population = logistic_growth(train_years - train_years[0], P0, r, K)
pred_test_population = logistic_growth(test_years - train_years[0], P0, r, K)

# Calculate metrics
train_mae = mean_absolute_error(train_population, pred_train_population)
train_mse = mean_squared_error(train_population, pred_train_population)
train_r2 = r2_score(train_population, pred_train_population)

test_mae = mean_absolute_error(test_population, pred_test_population)
test_mse = mean_squared_error(test_population, pred_test_population)
test_r2 = r2_score(test_population, pred_test_population)

# Print metrics
print(f'Train MAE: {train_mae:.2f}')
print(f'Train MSE: {train_mse:.2f}')
print(f'Train R²: {train_r2:.2f}')
print(f'Test MAE: {test_mae:.2f}')
print(f'Test MSE: {test_mse:.2f}')
print(f'Test R²: {test_r2:.2f}')


# Plot the actual data and the logistic growth forecast
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='--', label='Actual Population')
plt.plot(forecast_df_logistic['Year'], forecast_df_logistic['Forecasted Population'], marker='o', linestyle='--', color='red', label='Forecasted Population (Logistic Model)')
plt.plot(train_years, train_population, marker='o', linestyle='-', label='Train Population')
plt.plot(test_years, test_population, marker='o', linestyle='-', label='Test Population')

plt.title('Population Forecast Using Logistic Growth Model')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
