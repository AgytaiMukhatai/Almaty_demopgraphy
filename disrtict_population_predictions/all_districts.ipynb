```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_districts.csv')
df
```

```python
df['Population'] = df['Population'].astype(str).str.replace(' ', '').astype(int)
df
```

```python
for column in df.columns:
    df[column] = df[column].astype(str).str.replace(' ', '').str.replace('\xa0', '')
    df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64')


print(df)
```

```python
df.info()
```

```python
plt.figure(figsize=(20, 12))

line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':']

for column, marker, color, label, linestyle in zip(
    ['Almaly', 'Alatau', 'Auezov', 'Bostandyk', 'Zhetisu', 'Medeu', 'Nauryzbai', 'Turksib'],
    ['o', 'o', 's', 'D', '^', 'x', 'x', 'v'],
    ['b', 'b', 'g', 'r', 'y', 'm', 'm', 'c'],
    ['Almaly', 'Alatau', 'Auezov', 'Bostandyk', 'Zhetisu', 'Medeu', 'Nauryzbai', 'Turksib'],
    line_styles
):
    
    filtered_data = df[['Year', column]].dropna()
    plt.plot(filtered_data['Year'], filtered_data[column], marker=marker, linestyle=linestyle, color=color, label=label)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Almaty districts population')
plt.legend()
plt.grid(True)
plt.show()
```
