```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

all_features = df.columns.drop(['Population', 'Year'])


train_data = df[df['Year'] <= 2018]
test_data = df[(df['Year'] >= 2019) & (df['Year'] <= 2022)]

X_train = train_data[all_features]
y_train = train_data['Population']

X_test = test_data[all_features]
y_test = test_data['Population']


r_model_train = LinearRegression()
K_model_train = LinearRegression()

r_model_train.fit(X_train, y_train)
K_model_train.fit(X_train, y_train)

r_params_train = [r_model_train.intercept_] + list(r_model_train.coef_)
K_params_train = [K_model_train.intercept_] + list(K_model_train.coef_)


def compute_r_K_limited(X, r_params, K_params, r_min, r_max, K_min, K_max):
    r = np.clip(r_params[0] + sum(r_param * x for r_param, x in zip(r_params[1:], X)), r_min, r_max)
    K = np.clip(K_params[0] + sum(K_param * x for K_param, x in zip(K_params[1:], X)), K_min, K_max)
    return r, K

def logistic_growth(P, t, r, K):
    return r * P * (1 - P / K)


def euler_method_limited(P0, t, r_params, K_params, X, r_min, r_max, K_min, K_max):
    dt = t[1] - t[0]
    P = [P0]
    for i in range(1, len(t)):
        r, K = compute_r_K_limited(X, r_params, K_params, r_min, r_max, K_min, K_max)
        P_next = P[-1] + logistic_growth(P[-1], t[i-1], r, K) * dt
        P.append(P_next)
    return P


r_min, r_max = 0, 1
K_min, K_max = 1e5, 1e8  


t_corrected = np.linspace(0, len(test_data) - 1, len(test_data) * 250)


P0_test = y_train.iloc[-1]


population_forecast_corrected = euler_method_limited(P0_test, t_corrected, r_params_train, K_params_train, X_test.iloc[-1].values, r_min, r_max, K_min, K_max)


predicted_population_corrected = population_forecast_corrected[::len(t_corrected) // len(test_data)]


mse_corrected = mean_squared_error(y_test, predicted_population_corrected)
r2_corrected = r2_score(y_test, predicted_population_corrected)


years_train = np.arange(2009, 2023)
population_full_train = np.concatenate((y_train.values, predicted_population_corrected))


plt.figure(figsize=(10, 6))
plt.plot(years_train, population_full_train, marker='o', linestyle='-', color='b', label='  ()')
plt.plot(df['Year'], df['Population'], marker='x', linestyle='-', color='r', label=' ')
plt.xlabel('')
plt.ylabel('')
plt.title('    (    2018 , )')
plt.legend()
plt.grid(True)
plt.show()


mse_corrected, r2_corrected
```

```python
def geometric_growth(P0, r, t):
    return P0 * (1 + r) ** t


P0 = y_train.iloc[0]
r_values = []


for i in range(1, len(y_train)):
    r = (y_train.iloc[i] / y_train.iloc[i-1]) - 1
    r_values.append(r)

r_mean = np.mean(r_values)


years_test = np.arange(2019, 2023) - 2018
predicted_population_geometric = geometric_growth(y_train.iloc[-1], r_mean, years_test)


mse_geometric = mean_squared_error(y_test, predicted_population_geometric)
r2_geometric = r2_score(y_test, predicted_population_geometric)


years_full = np.arange(2009, 2023)
population_full_geometric = np.concatenate((y_train.values, predicted_population_geometric))


plt.figure(figsize=(10, 6))
plt.plot(years_full, population_full_geometric, marker='o', linestyle='-', color='b', label='  (  )')
plt.plot(df['Year'], df['Population'], marker='x', linestyle='-', color='r', label=' ')
plt.xlabel('')
plt.ylabel('')
plt.title('    (    2018 ,   )')
plt.legend()
plt.grid(True)
plt.show()


mse_geometric, r2_geometric
```

```python
def verhulst_model(P0, r, K, t):
    return K * P0 * np.exp(r * t) / (K + P0 * (np.exp(r * t) - 1))


P0 = y_train.iloc[0]
r_values = []


for i in range(1, len(y_train)):
    r = np.log(y_train.iloc[i] / y_train.iloc[i-1])
    r_values.append(r)

r_mean = np.mean(r_values)


K = y_train.max() * 1.5  


years_test = np.arange(2019, 2023) - 2018
predicted_population_verhulst = verhulst_model(y_train.iloc[-1], r_mean, K, years_test)


mse_verhulst = mean_squared_error(y_test, predicted_population_verhulst)
r2_verhulst = r2_score(y_test, predicted_population_verhulst)


years_full = np.arange(2009, 2023)
population_full_verhulst = np.concatenate((y_train.values, predicted_population_verhulst))


plt.figure(figsize=(10, 6))
plt.plot(years_full, population_full_verhulst, marker='o', linestyle='-', color='b', label='  (  )')
plt.plot(df['Year'], df['Population'], marker='x', linestyle='-', color='r', label=' ')
plt.xlabel('')
plt.ylabel('')
plt.title('    (    2018 ,   )')
plt.legend()
plt.grid(True)
plt.show()


mse_verhulst, r2_verhulst
```

```python
from scipy.integrate import odeint
def lotka_volterra_model(Y, t, alpha, beta, delta, gamma):
    P1, P2 = Y
    dP1_dt = alpha * P1 - beta * P1 * P2
    dP2_dt = delta * P1 * P2 - gamma * P2
    return [dP1_dt, dP2_dt]


P0 = y_train.iloc[0]
P2_0 = y_train.mean()  


alpha = 0.1
beta = 0.02
delta = 0.01
gamma = 0.1


t = np.linspace(0, len(test_data), len(test_data))


initial_conditions = [P0, P2_0]
solution = odeint(lotka_volterra_model, initial_conditions, t, args=(alpha, beta, delta, gamma))
P1_solution = solution[:, 0]


mse_lv = mean_squared_error(y_test, P1_solution)
r2_lv = r2_score(y_test, P1_solution)


years_full = np.arange(2009, 2023)
population_full_lv = np.concatenate((y_train.values, P1_solution))


plt.figure(figsize=(10, 6))
plt.plot(years_full, population_full_lv, marker='o', linestyle='-', color='b', label='  ( -)')
plt.plot(df['Year'], df['Population'], marker='x', linestyle='-', color='r', label=' ')
plt.xlabel('')
plt.ylabel('')
plt.title('    (    2018 ,  -)')
plt.legend()
plt.grid(True)
plt.show()

mse_lv, r2_lv
```
