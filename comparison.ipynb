```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('demography_cleaned.csv')
df
```

```python
pd.set_option('display.max_columns', None)
print("Initial DataFrame:")
df
```

```python
# Splitting the data into training and testing sets
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
train_data = df[df['Year'] <= 2018]
test_data = df[(df['Year'] >= 2019) & (df['Year'] <= 2022)]

y_train = train_data['Population']
y_test = test_data['Population']

# Define the Malthusian model function
def malthusian_model(P0, r, t):
    return P0 * np.exp(r * t)

# Calculate the initial population and r values
P0 = y_train.iloc[0]
r_values = []

for i in range(1, len(y_train)):
    r = np.log(y_train.iloc[i] / y_train.iloc[i-1])
    r_values.append(r)

r_mean = np.mean(r_values)

# Predicting the population up to 2035
years_forecast = np.arange(2019, 2036) - 2018
predicted_population_malthus_forecast = malthusian_model(y_train.iloc[-1], r_mean, years_forecast)

# Combining the actual and forecasted populations
years_full_forecast = np.arange(2009, 2036)
population_full_malthus_forecast = np.concatenate((y_train.values, predicted_population_malthus_forecast))

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(years_full_forecast, population_full_malthus_forecast, marker='o', linestyle='-', color='b', label='Predicted Population (Malthusian Model)')
plt.plot(df['Year'], df['Population'], marker='x', linestyle='-', color='r', label='Actual Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Predicted and Actual Population (Using Data Until 2018, Malthusian Model)')
plt.legend()
plt.grid(True)
plt.show()

# Calculating MSE and R2 scores for the forecasted period (2019-2023)
predicted_population_malthus = malthusian_model(y_train.iloc[-1], r_mean, np.arange(2019, 2023) - 2018)
mse_malthus = mean_squared_error(y_test, predicted_population_malthus)
r2_malthus = r2_score(y_test, predicted_population_malthus)

mse_malthus, r2_malthus
```
